# 04.02 - Simple Scatter Plots

This section is covering the simple scatter plot.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

### Scatter Plots with <code>plt.plot()</code>

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='green');
```

There are several types of symbols we can use for the plotting. Here are some of the most basic types:

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.4);
```

```python
plt.plot(x, y, '-ok');
```

Additional properties of the graph can be specified:

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

### Scatter Plots with <code>plt.scatter</code>

```python
plt.scatter(x, y, marker='>');
```
