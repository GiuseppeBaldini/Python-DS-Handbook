# 04.03 - Visualizing Errors

An essential element of describing measurements is defining errors, and Matplotlib is a great tool to visualize them:

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

### Basic Errorbars

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

Here <code>fmt</code> is controlling the format of lines and points. The <code>errorbar</code> can also be customized, for example making it ligther for crowded plots.

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```
